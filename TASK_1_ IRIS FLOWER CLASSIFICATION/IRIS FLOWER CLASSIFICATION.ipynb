#  IRIS FLOWER CLASSIFICATION ðŸŒ¿

Iris flower has three species; setosa, versicolor, and virginica, which differs according to theirmeasurements.    
To train a machine learning model that can learn from the measurements of the iris species and classify them.

```python
# Importing libraries
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
%matplotlib inline
```

## DataSet Loading

```python
df = pd.read_csv("Iris.csv")
df.head(5)
```

## Data Visualiztion

```python
#Visualize the whole dataset
sns.pairplot(df)
```

## Data Preprocessing

```python
# Removing the id column as it is of no use
df.drop(['Id'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# Shuffling the datset to improve learning 
df.sample(frac=1)
```

```python
df.describe()
```

```python
# Seperating feature and target
data = df.values

x=data[:,0:4]
y=data[:,4]

print(y)
```

```python
# Splitting the data into Training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

## Model Fitting and predicting

```python
# Support Vector Machine
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(x_train,y_train)
```

```python
pred1 = model_svc.predict(x_test)
# For accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pred1)*100)
```

```python
for i in range(len(pred1)):
    print(y_test[i],pred1[i])
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression()
model_lr.fit(x_train,y_train)
```

```python
pred2 = model_lr.predict(x_test)
# For accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pred2)*100)
```

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
model_dtc = DecisionTreeClassifier()
model_dtc.fit(x_train,y_train)
```

```python
pred3 = model_dtc.predict(x_test)
# For accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pred3)*100)
```

```python
# Classificcation Report
from sklearn.metrics import classification_report
print(classification_report(y_test,pred1))
```

```python
x_check = np.array([[3,3,1.76,1.98],[4.8,9.7,9.6,8.5],[3.4,3.4,1.4,1.4]])
# Custom prediction

pred = model_svc.predict(x_check)
print("The species of the iris flower is", pred)
```
